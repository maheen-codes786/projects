```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Example usage:
review = "This app is amazing!"
sentiment_scores = sia.polarity_scores(review)

# Interpret the sentiment scores
if sentiment_scores['compound'] >= 0.05:
    sentiment = "Positive"
elif sentiment_scores['compound'] <= -0.05:
    sentiment = "Negative"
else:
    sentiment = "Neutral"

print(f"Review: {review}")
print(f"Sentiment: {sentiment}")
```

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Sample data (replace with your data)
data = {'Review': ["This app is great!",
                  "Terrible experience, waste of time.",
                  "It's okay, not too bad."]}
df = pd.DataFrame(data)

# Perform sentiment analysis and add the results to the DataFrame
df['Sentiment'] = df['Review'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Interpret the sentiment scores
df['Sentiment_Label'] = df['Sentiment'].apply(lambda x: "Positive" if x >= 0.05 else ("Negative" if x <= -0.05 else "Neutral"))

# Display the DataFrame with sentiment analysis results
print(df)
```
